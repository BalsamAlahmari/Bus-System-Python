```python


 # Define a function to print student details
def student_to_uni(student_id,shift):
    filtered_df = data_bus[(data_bus['STUDENT_ID'] == student_id)& (data_bus['WORK_TIME'] == shift)]
    if not filtered_df.empty:
        student_info = filtered_df.iloc[0]
        print("****************************Driver And Bus Info******************************")
        print(f"TO: {student_info['TO_UNI']}")
        print("-----------------------------------------------------------------------------")
        print(f"The Bus Is Available By {student_info['time_to_uni']}. It Will Wait For 15 Minutes Before Leaving")
        print("-----------------------------------------------------------------------------")
        print(f"Driver Name: {student_info['D_FNAME']}")
        print("-----------------------------------------------------------------------------")
        print(f"Bus Number: {student_info['BUS_NUMBER']}")
        print("-----------------------------------------------------------------------------")
        print(f"Driver Phone: 0{student_info['D_PHNUMBER']}")
        print("-----------------------------------------------------------------------------")
    else:
        print("-----------------------------------------------------------------------------")
        print("You are not registered in our database.")

def student_to_home(student_id,shift):
    filtered_df = data_bus[(data_bus['STUDENT_ID'] == student_id)& (data_bus['WORK_TIME'] == shift)]
    if not filtered_df.empty:
        student_info = filtered_df.iloc[0]
        print("****************************Driver And Bus Info******************************")
        print(f"TO : {student_info['B_STREET']},{student_info['S_POSTCODE']}")
        print("-----------------------------------------------------------------------------")
        print(f"The Bus Is Available By {student_info['time_to_home']}. It Will Wait For 15 Minutes Before Leaving")
        print("-----------------------------------------------------------------------------")
        print(f"Driver Name: {student_info['D_FNAME']}")
        print("-----------------------------------------------------------------------------")
        print(f"Bus Number: {student_info['BUS_NUMBER']}")
        print("-----------------------------------------------------------------------------")
        print(f"Driver Phone: 0{student_info['D_PHNUMBER']}")
        print("-----------------------------------------------------------------------------")
    else:
        print("-----------------------------------------------------------------------------")
        print("You are not registered in our database.")
```

```python
while True:
    try:
        print("*****************************Student Info************************************")
        student_id_input=int(input('please enter your ID? '))
        break
    except ValueError:
        print("Invalid student ID. Please enter an integer.")
        continue

valid_shifts = ("morning", "afternoon")
while True:
    print("-----------------------------------------------------------------------------")
    student_shift = input("Morning shift OR Afternoon shift? ").strip()
    student_shift = student_shift.lower()
    if student_shift not in valid_shifts:
        print("Invalid student shift. Please enter 'Morning' or 'Afternoon'.")
        continue
    if student_shift =='morning':
        student_shift="Morning Shift"
    if student_shift =='afternoon':
        student_shift="Afternoon Shift"
    break
    
valid_des = ("uni","home")
while True:
    print("-----------------------------------------------------------------------------")
    student_destination=input('To home OR To uni? ')
    student_destination = student_destination.lower()
    if student_destination not in valid_des:
        print("Invalid student destination. Please enter 'UNI or 'HOME'.")
        continue
    break

    
    
if  student_destination=='uni':
    student_to_uni( student_id_input,student_shift)
elif student_destination=='home':
     student_to_home( student_id_input,student_shift)
```
